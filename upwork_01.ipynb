```python
import numpy as np
import pandas as pd
import openpyxl
import re
```

```python
df_accounts = pd.read_excel(r'./SampleAccounts.xlsx')
df_users = pd.read_csv(r'./SampleUsers.csv')
```

```python
df_accounts = df_accounts.drop(df_accounts[df_accounts['CUSTOMER_CLASS']=='Commercial'].index)
# dropping commercial accounts
```

```python
df_accounts['name_with_middle_and_address'] = df_accounts["ENTITY_NAME"]
```

```python
df_accounts['name_without_middle_and_address'] = df_accounts["ENTITY_NAME"]
for indexx, entry in df_accounts.iterrows():
    #print(index, entry)
    if " " in entry["name_without_middle_and_address"]:
        location = entry["name_without_middle_and_address"].index(" ")
        cutname = entry["name_without_middle_and_address"][0:location]
        df_accounts['name_without_middle_and_address'][indexx] = cutname
```

```python
columnlist = ["name_with_middle_and_address", "name_without_middle_and_address"]
df=df_accounts
for column in columnlist:
    df[column] = df[column]+df["ADDRESS1"]+df["CITY"]++df["STATE"]
    df[column] = df[column].str.replace(" ","")
    df[column] = df[column].str.replace(",","")
    df[column] = df[column].str.upper()
#print (type(df_accounts['name_with_middle_and_address'][2]))
#print ((df_accounts['name_with_middle_and_address'][2]))
df_accounts.head()
#df_users.head()
```

```python
df_users["name_without_middle_and_address"] = ""
df_users["name_with_middle_and_address"] = ""

for indexx, entry in df_users.iterrows():
    #print(indexx, entry["Middle_Name"], type(entry["Middle_Name"]))
    if type(entry["Middle_Name"]) == float:
        #print("found float")
        fullname = entry["Last_Name"] + entry ["First_Name"]
        df_users["name_with_middle_and_address"][indexx] = fullname
        df_users["name_without_middle_and_address"][indexx] = fullname
    else:
        fullname = entry["Last_Name"] + entry ["First_Name"] + entry ["Middle_Name"]
        df_users["name_with_middle_and_address"][indexx] = fullname
        df_users["name_without_middle_and_address"][indexx] = entry["Last_Name"] + entry ["First_Name"]
df_users.head()
```

```python
columnlist = ["name_with_middle_and_address", "name_without_middle_and_address"]
df=df_users
for column in columnlist:
    df[column] = df[column]+df["Formatted_Address"]+df["City_State"]
    df[column] = df[column].str.replace(" ","")
    df[column] = df[column].str.replace(",","")
    df[column] = df[column].str.upper()
df_users.head()
```

```python
new_column_names = ["FirstName", "LastName", "MiddleName", "Address1", "Address2", "City", "State", "ZIP", "County"]
df_MatchRecords = pd.DataFrame(columns = new_column_names)
df_NewRecords = pd.DataFrame(columns = new_column_names)

col1 = "name_with_middle_and_address"
col2 = "name_without_middle_and_address"

for index1, record1 in df_accounts.iterrows():
    
    fullname = df_accounts["ENTITY_NAME"][index1]
    fullname_split = re.split(',| ', fullname)
    # print(fullname_split)
    middlename = ""
    if len(fullname_split)>2 : middlename = fullname_split[-1]
    lastname = fullname_split[0]
    firstname = fullname_split[1]

    for index2, record2 in df_users.iterrows():
        
        
        
        if ( df_accounts[col1][index1] == df_users[col1][index2] ) or ( df_accounts[col2][index1] == df_users[col2][index2] ):
            matched = True
            
            to_append = [firstname, lastname, middlename,
                         df_accounts["ADDRESS1"][index1], df_users["Formatted_Address"][index2], df_accounts["CITY"][index1], 
                         df_accounts["STATE"][index1], df_accounts["POSTAL"][index1], df_accounts["COUNTY"][index1] ]
            df_length = len(df_MatchRecords)
            df_MatchRecords.loc[df_length] = to_append
            break
        else:
            matched = False
    if not matched:
        to_append = [firstname, lastname, middlename,
                     df_accounts["ADDRESS1"][index1], "", df_accounts["CITY"][index1], 
                     df_accounts["STATE"][index1], df_accounts["POSTAL"][index1], df_accounts["COUNTY"][index1] ]
        df_length = len(df_NewRecords)
        df_NewRecords.loc[df_length] = to_append
        
df_MatchRecords.to_excel("MatchRecords.xlsx")  
#df_MatchRecords.head()
```
